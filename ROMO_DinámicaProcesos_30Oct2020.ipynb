```
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```
def EHST(x,t):
  #variables dependientes
  x1=x[0]
  x2=x[1]

  #
  m_w = 10  #min
  mc_ha = 1.0 #min
  mc_wc= 1.0 #min
  uno_wc = 0.05 #C min/kcal
  Q = 5000  #Kcal/min
  T = 120 #C

  #definiendo términos grandes
  a = m_w*mc_ha
  b = mc_ha + mc_wc + m_w
  c = T + uno_wc*Q
  

  #cambios de variables
  #x1' = T'
  #x2' = T''

  dx2dt = (c - x1 - b*x2 ) / a
  dx1dt = x2

  return [dx1dt,dx2dt]
```

```
xo = [350,0]  #valores iniciales

t_0 = 0   #tiempo valor inicial
t_f = 100   #tiempo valor final
pun =  100   #cantidad puntos
t= np.linspace(t_0,t_f,pun)   #vector tiempo

#resolviendo ODEs
y=odeint(EHST,xo,t)
```

```
plt.plot(t,y[:,0] , "r--")
plt.xlabel("Tiempo [min] ")
plt.ylabel( "Temperatura °C")
plt.grid()
plt.show()
plt.plot(t,y[:,1] , "b.-")
plt.xlabel("Tiempo [min] ")
plt.ylabel( "Derivada Temperatura °C")
plt.grid()
plt.show()
```
