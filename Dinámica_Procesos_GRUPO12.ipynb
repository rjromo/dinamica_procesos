```
#importamos librerías

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
#cargando excel

data= pd.read_excel("/content/dataCL02.xlsx").values
dimensiones = np.shape(data)
#imprimiendo dimensiones de la matriz
dimensiones
```

```
#creando vector linealmente independiente
t = np.linspace(1,30,30)


```

```
#creando ciclo for
for i in range(dimensiones[1]):
  if i % 2 != 0:  #esto imprime los pares de NUESTRO EXCEL
    plt.plot(t,data[:,i],color='yellow', linestyle='dashed')
    plt.xlabel("Tiempo")
    plt.ylabel("Datos grupos {}".format(i))
  plt.show()
```
